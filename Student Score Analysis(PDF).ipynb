```python
import pandas as pd
import pdfplumber
```

```python
pdf_path = r"D:\Python\Student Mark.pdf"
```

```python
rows = []
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            rows.extend(table)
            
df = pd.DataFrame(rows[1:], columns=rows[0])
```

# Top Ranking Students

```python
print("Top Performers:")
print(df.head())
```

# Low Ranking Students

```python
print("\nLeast Performers:")
print(df.tail())
```
